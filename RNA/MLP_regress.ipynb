<h1>Rede Neural para Regressão</h1> 
Dataset "Advertising"

Importando módulos e bibliotecas:

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

Lendo dados do arquivo CSV:

```python
df = pd.read_csv('data/advertising.csv')
df.head()
```

Dividindo em treinamento (80%) e teste (20%):

```python
x_val = df.drop(['Sales'], axis=1)
y_val = df['Sales']

X_train, X_test, y_train, y_test = train_test_split(x_val, y_val, test_size=0.2, random_state=42)
```

Plotando dados.

```python
plt.scatter(X_train['TV'], y_train, label='TV Ad Budget')
plt.scatter(X_train['Radio'], y_train, label='Radio Ad Budget')
plt.scatter(X_train['Newspaper'], y_train, label='Newspaper Ad Budget')
plt.xlabel('Advertising Budget ($)')
plt.ylabel('Revenue ($)')
plt.legend()
plt.show()
```

Normalizando os dados

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train) # Don't cheat - fit only on training data
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test) # apply same transformation to test data
```

Criando e utilizando MLP para regressão

```python
from sklearn.neural_network import MLPRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

model = MLPRegressor(
    hidden_layer_sizes=[62,62], 
    activation='relu',
    solver='adam',
    verbose=True,
    random_state=1,
    max_iter=1400)
model = model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
model.score(X_test, y_test) # R Squared
```

```python
plt.scatter(X_test[:,0], y_test, color='blue')
plt.scatter(X_test[:,0], y_pred, color='red')

# Set the axis labels and title
plt.xlabel('TV Ad Budget ($)')
plt.ylabel('Sales ($)')
plt.title('Neural Network Regression - TV Ad Budget vs Sales')

# Add a legend to the plot
plt.legend(['True Values', 'Predictions'])
plt.show()
```

```python
plt.scatter(X_test[:,1], y_test, color='blue')
plt.scatter(X_test[:,1], y_pred, color='red')

# Set the axis labels and title
plt.xlabel('Radio Ad Budget ($)')
plt.ylabel('Sales ($)')
plt.title('Neural Network Regression - Radio Ad Budget vs Sales')

# Add a legend to the plot
plt.legend(['True Values', 'Predictions'])
plt.show()
```

<h3>Imprimindo métricas do modelo</h3>

```python
from sklearn import metrics
import numpy as np

def print_evaluate(real, predicted):
    mae = metrics.mean_absolute_error(real, predicted)
    mape = metrics.mean_absolute_percentage_error(real, predicted)
    mse = metrics.mean_squared_error(real, predicted)
    rmse = np.sqrt(mse)
    r2_square = metrics.r2_score(real, predicted)
    
    print("MAE:", mae)
    print("MAPE:", mape)
    print("MSE:", mse)
    print("RMSE:", rmse)
    print("R2 Square:", r2_square)
```

```python
print("Métricas de Desempenho (RNA):")
print("\nDados de teste:")
print_evaluate(y_test, y_pred)
```

<h3>Comparando com regressor linear</h3>

```python
from sklearn.linear_model import LinearRegression

# regressão (treinamento)
my_model = LinearRegression()
my_model.fit(X_train, y_train)
print("R2 (test) =", my_model.score(X_test, y_test)) # r2 squared
```

```python
#regressão (teste)
y_pred = my_model.predict(X_test)
print("Métricas de Desempenho (Linear):")
print("\nDados de teste:")
print_evaluate(y_test, y_pred)
```

<h3>Comparando com SVM</h3>

```python
from sklearn.svm import SVR
svr = SVR(C=30.0, epsilon=0.5)
svr.fit(X_train, y_train)
svr.score(X_test, y_test)

#regressão (teste)
y_pred = svr.predict(X_test)
print("Métricas de Desempenho (Linear):")
print("\nDados de teste:")
print_evaluate(y_test, y_pred)
```
